```python
#W6A9
import ast
d1 = ast.literal_eval(input())
d2 = ast.literal_eval(input())
d = d1.copy()
for k in d2:
    if k in d:
        d[k] += d2[k]
    else:
        d[k] = d2[k]
s = sorted(d.keys())
r = {}
for k in s:
    r[k] = d[k]
print(r)
```

```python
#W6A10
s = input()
a = []
for x in s:
    a.append(int(x))
k = int(input())
r = []
for i in range(len(a)):
    for j in range(i + 1, len(a)):
        if a[i] + a[j] == k:
            r.append((a[i], a[j]))
r.sort()
print(r)
```

```python
#W6A12
a = list(map(int, input().split()))
d = {}
for x in a:
    d[x] = d.get(x, 0) + 1
m = 0
for x in d:
    if d[x] > m:
        m = d[x]
r = None
for x in d:
    if d[x] == m:
        if r is None or x < r:
            r = x
print(r)
```

```python
#W6A19
def f(a):
    if not a: return -1
    m = a[0]
    p = 0
    for i in range(1, len(a)):
        if a[i] > m:
            m = a[i]
            p = i
    return p

a = list(map(int, input().split()))
print(f(a))
```

```python
#W6A20
def f(a, k):
    p = -1
    for i in range(len(a)):
        if a[i] == k:
            p = i
            break
    return p

a = list(map(int, input().split()))
k = int(input())
print(f(a, k))
```
