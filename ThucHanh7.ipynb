```python
#w7a1
def binary_search(arr, left, right, target):
    a = arr
    l, r, val = left, right, target
    l, r = 0, int(len(a)-1)
    while(l <= r):
        mid = (l + r) // 2
        if a[mid] == val:
            return mid
        if a[mid] < val:
            l = mid + 1
        else:
            r = mid - 1
    return -1 

a = list(map(int, input().split()))
val = int(input())
print(binary_search(a, 0, len(a)-1, val))
```

```python
#w7a3
def swap(a: int, b: int):
    a = b - a
    b = b - a
    a = b + a
    return (a, b)

def bubble_sort(arr: list, n: int):
    for i in range(n):
        for j in range(n-i-1):
            if a[j] > a[j+1]:
                a[j], a[j+1] = swap(a[j], a[j+1])
    return arr

def selection_sort(arr: list, n: int):
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if a[j] < a[min_idx]:
                min_idx = j
                
        a[i], a[min_idx] = a[min_idx], a[i]
        
    return arr

a = list(map(int, input().split()))
# a = [9, 8, 5, 1, 2, 4, 6]
n = len(a)
a = bubble_sort(a, n)
b = selection_sort(a, n)
if a == b:
    print(a)
```

```python
#w7a4
import heapq
a = list(map(int, input().split()))
n = len(a)
cnt = {}
pq = []
for i in range(n):
    if a[i] not in cnt:
        cnt[a[i]] = [-1, i]
    else:
        cnt[a[i]][0] -= 1
        cnt[a[i]][1] = i
heapq.heapify(pq)
for key, value in cnt.items():
    heapq.heappush(pq, [value[0], value[1], key])
ans = heapq.heappop(pq)
print(ans[2], 'xuat hien nhieu nhat, som nhat,', -ans[0], 'lan')
```

```python
#w7a7
def lower_bound(a, x):
    n = len(a)
    l, r = 0, n
    while l < r:
        mid = (l+r)//2
        if a[mid] >= x:
            r = mid
        else:
            l = mid + 1
    return l

s = input()
a = []
tmp = ""

for i in s:
    if i.isdigit() or i == '-':
        tmp += i
    else:
        if tmp:
            a.append(int(tmp))
            tmp = ""
            
tail = []
for i in a:
    if len(tail) == 0:
        tail.append(i)
        continue
    idx = lower_bound(tail, i)
    if idx < len(tail):
        tail[idx] = i
    else:
        tail.append(i)
print(len(tail))            
```

```python
#w7a8

def find_l(a, x):
    l, r = 0, len(a) - 1
    res = len(a)
    while l <= r:
        mid = (l + r) // 2
        if a[mid] >= x:
            res = mid
            r = mid - 1
        else:
            l = mid + 1
    return res

def find_r(a, y):
    l, r = 0, len(a) - 1
    res = -1
    while l <= r:
        mid = (l + r) // 2
        if a[mid] <= y:
            res = mid
            l = mid + 1
        else:
            r = mid - 1
    return res

def get_input(s):
    a = []
    tmp = ""
    for char in s:
        if char.isdigit() or char == '-': 
            tmp += char
        else:
            if tmp:
                a.append(int(tmp))
                tmp = ""
    if tmp: 
        a.append(int(tmp))
    return a

def find_next(nxt, v):
    if v == nxt[v]:
        return v
    nxt[v] = find_next(nxt, nxt[v])
    return nxt[v]

# s = "[[2, 8], [1, 5], [10, 15]]"
# s1 = "[1, 4, 6, 11, 20]"
s = input()
s1 = input()

a = get_input(s)
q = get_input(s1)

b = []
n = len(a)
for i in range(0, n, 2):
    tmp = (a[i+1] - a[i] + 1, a[i], a[i+1])
    b.append(tmp)

b.sort()

q1 = []
for i in range(len(q)):
    q1.append((q[i], i))
q1.sort() 

q2 = [x[0] for x in q1] 
ans = [-1] * len(q)

nxt = list(range(len(q2) + 1))

for length, x, y in b:
    left = find_l(q2, x)
    right = find_r(q2, y)
    
    i = find_next(nxt, left)
    while i <= right:
        ans[q1[i][1]] = length
        
        nxt[i] = find_next(nxt, i + 1)
        i = nxt[i]
print(ans)
```

```python
#w7a10
def lower_bound(a, x):
    l, r = 0, len(a)
    while l < r:
        mid = (l + r) // 2
        if a[mid] >= x:
            r = mid
        else:
            l = mid + 1
    return l

def get_input(s):
    a = []
    tmp = ""
    for char in s:
        if '0' <= char <= '9' or char == '-':
            tmp += char
        else:
            if tmp:
                a.append(int(tmp))
                tmp = ""
    if tmp:
        a.append(int(tmp))
    return a

s_nums = input()
s_m = input()

nums = get_input(s_nums)
m_val = get_input(s_m)[0]

max_sum = 0
current_prefix_sum = 0
seen_prefix_sums = [0] 

for x in nums:
    current_prefix_sum = (current_prefix_sum + x) % m_val
    max_sum = max(max_sum, current_prefix_sum)
    
    idx = lower_bound(seen_prefix_sums, current_prefix_sum + 1)
    
    if idx < len(seen_prefix_sums):
        val = (current_prefix_sum - seen_prefix_sums[idx] + m_val) % m_val
        max_sum = max(max_sum, val)
    
    pos = lower_bound(seen_prefix_sums, current_prefix_sum)
    seen_prefix_sums.insert(pos, current_prefix_sum)

print(max_sum)
```
