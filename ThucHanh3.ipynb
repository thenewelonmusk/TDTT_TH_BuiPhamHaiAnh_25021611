```python
# Phân loại tam giác
def tgcan(a, b, c):
    if a == b == c:
        return 'Tam giác đều'
    elif a == b or a == c or b == c:
        return 'Tam giác cân'
    else:
        return 'Tam giác thường'
a,b,c = map(int, input().split())
if a + b > c and b + c > a and a + c > b:
    print(tgcan(a, b, c))
else:
    print('Không phải tam giác')
```

```python
# Tính tiền điện
n = int(input())
ans = 0
if n >= 100:
   ans += (n - 100) * 3
   n = 100
if n >= 51:
   ans += (n - 50) * 2
   n = 50
if n >= 0:
   ans += (n - 0) * 1.5
print(int(ans*1000))
```

```python
# Rounding
def absolute(n):
    if n < 0:
        n = -n
    return n
def up(n):
    if n % 1 == 0:
        return n
    ans = 0
    if n >= 0:
        ans = n + (1-(n % 1))
    else:
        ans = n + (1 - (n % 1))
    return ans 
def down(n):
    if n % 1 == 0:
        return n
    ans = 0
    if n >= 0:
        ans = n - (n % 1)
    else:
        ans = n - (n % 1)
    return ans
def nearest(n):
    a = down(n)
    b = up(n)
    dif1 = absolute(n - a)
    dif2 = absolute(n - b) 
    if(dif1 <= dif2):
        return a
    else:
        return b
n = float(input())
print(f'{up(n):.0f}', f'{down(n):.0f}', f'{nearest(n):.0f}', sep=' ')
```
