```python
#W4A2 PrimeNumber
def str_to_int(num: str) -> int:
    try:
        return int(num)
    except ValueError:
        return None

def check_positive_int(num):
    tmp = str_to_int(num)
    if tmp != None and tmp > 0:
        return tmp
    return 0

def is_prime(n: int):
    i = 2
    n = int(n)
    # print(type(i), type(n))
    while(int(i*i) <= n):
        if(n % i == 0):
            return False
        i = i + 1
    return True

n = input()
while(check_positive_int(n) == 0):
    n = input()

print(is_prime(n))
```

```python
#W4A6 SumPrimesInRange

def is_prime(n: int) -> bool:
    i = 2
    while(i*i <= n):
        if(n % i == 0):
            return False
        i += 1
    return True

def solve(a: int, b: int) -> int:
    ans =0 
    for i in range(a, b+1):
        if(is_prime(i)):
            ans += i
    return ans

a, b = map(int, input().split())
print(solve(a, b))
```

```python
#W4A7 LargestPrimeFactor

def is_prime(n: int) -> bool:
    i = 2
    while(i*i <= n):
        if(n % i == 0):
            return False
        i += 1
    return True

def solve(n: int):
    if is_prime(n):
        return n
    for i in range(n//2, 1, -1):
        if n % i == 0 and is_prime(i):
            return i

n = int(input())

print(solve(n))
```

```python
#W4A8 ReverseAddPalindrome

def rev(n: int):
    tmp = str(n)
    tmp = tmp[::-1]
    tmp = int(tmp)
    return tmp

def is_palin(tmp:int) -> bool:
    tmp = str(tmp)
    n = len(tmp)
    for i in range(0, n//2+1):
        if tmp[i] != tmp[n-i-1]:
            return False
    return True

n = int(input())
cnt = 0
while not is_palin(n):
    cnt += 1
    n = n + rev(n)
print(cnt, n)
```

```python
#W4A9 PerfectSquaresDistinctDigits

def is_square(n: int):
    if n < 0:
        return False
    tmp = int(n ** 0.5)

    return tmp * tmp == n

def valid(n: int):
	if not is_square(n):
		return False
	s = str(n)
	return len(s) == len(set(s))

n = int(input())
a = []
for i in range(1, n+1):
    if valid(i):
        a.append(i)

if len(a) == 0:
    print('no number')
else:
    for i in a:
        print(i, end=' ')
```

```python
#W4A10 Longest CollatzUpToN

def calc(n: int):
    cnt = 0
    while n != 1:
        # print(n)
        cnt += 1
        if n % 2 == 0:
            n = int(n / 2)
        else:
            n = n*3+1
    return cnt + 1

n = int(input())

mx = mx_cnt = 0

for i in range(1, n+1):
    cnt = calc(i)
    if cnt > mx_cnt:
        mx = i
        mx_cnt = cnt
        
print(mx, mx_cnt)
```

```python
#W4A13 ClosePairNumber

def sum_divisor(n: int):
    tmp = n
    ans = 1
    i = 2
    while i*i <= tmp:
        cur = 0
        while n % i == 0:
            n = int(n / i)
            cur = cur + 1
        if cur != 0:
            x = ((i ** (cur + 1) - 1) / (i - 1))
            ans = ans * x
        i = i + 1
    if n != 1:
        ans = ans * int(((n ** 2 - 1) / (n-1)))
    return int(ans - tmp)

a, b = map(int, input().split())

a1, b1 = sum_divisor(a), sum_divisor(b)
if a1 == b and b1 == a:
    print('true')
else:
    print('false') 
```

```python
#W4A32
def is_valid(s):
    if len(s) <= 6:
        return False

    special = "!@#$%^&*()-_+=[]{}|;:,.<>?/~`"
    i1 = i2 = i3 = i4 = False
    for i in s:
        if i in special:
            i1 = True
        elif i.isupper():
            i2 = True
        elif i.islower():
            i3 = True
        elif i.isdigit():
            i4 = True
    return (i1 and i2 and i3 and i4)

s = input()
print(is_valid(s))
```
