```python
#W5A1
def Max_Of_Two(a, b) -> int:
    return max(a, b)

a, b = map(int, input().split())
print(Max_Of_Two(a, b))
```

```python
#W5A2
def swap(a, b) -> tuple[int, int]:
    res = (b, a)
    return res

a, b = map(int, input().split())
ans = swap(a, b)
print(ans[0], ans[1])
```

```python
#W5A3
def is_prime(n):
	if n <= 1:
		return False
	if n == 2:
		return True
	if n % 2 == 0:
		return False

	limit = int(n**0.5) + 1
	for i in range(3, limit, 2):
		if n % i == 0:
			return False
	
	return True
		
n = int(input())
print(is_prime(n))
```

```python
#W5A4
def is_valid(n:int):
    res = 1
    for i in range(2, int(n**0.5)+1, 1):
        if n % i == 0:
            res += i + (n / i)
    return n == int(res)

n = int(input())
print(is_valid(n))
```

```python
#W5A5
def solve(a: list, k: int):
    for i in range(len(a)):
        if a[i] == k:
            return i+1
    return -1

input_str = input()
a = list(map(int, input_str.split()))
k = int(input())
print(solve(a, k))
```

```python
#W5A6
def fac(n):
    ans = 1
    for i in range(1, n+1):
        ans = ans * i
    return ans

n = int(input())
print(fac(n))
```

```python
#W5A7

def calc(a, b, op):
    if op == '+':
        return a + b
    if op == '/':
        return a / b
    if op == '*':
        return a * b
    if op == '-':
        return a * b

a, op, b = input().split()
a = float(a)
b = float(b)
print(f"{calc(a, b, op):.2f}")
```

```python
#W5A8
def solve(a, b):
    ans = a ^ b
    return ans.bit_count()

a, b = map(int, input().split())
print(solve(a,b))
```

```python
#W5A9

def solve(n):
    ans = 0
    for i in n:
        ans += int(i)
    return ans

n = input()
print(solve(n))
```

```python
#W5A10

def valid(s, s1, a, b):
    for i in range(len(a)):
        if a[i] not in s:
            s[a[i]] = b[i]
        else:
            if s[a[i]] != b[i]:
                return 'false'
    for i in range(len(b)):
        if b[i] not in s1:
            s1[b[i]] = a[i]
        else:
            if s1[b[i]] != a[i]:
                return 'false'
    return 'true'
        

a, b = input().split()
s = s1 = {}
print(valid(s, s1, a, b))
```
