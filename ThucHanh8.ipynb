```python
#w8a1
class Cylinder:
    def __init__(self, r, h):
        self.r = r
        self.h = h

    def area(self):
        return 2 * 3.14 * self.r * (self.r + self.h)

    def vol(self):
        return 3.14 * self.r**2 * self.h

r, h = map(float, input().split())
c = Cylinder(r, h)
print(f"{c.area():.2f} {c.vol():.2f}")
```

```python
#w8a2
class Date:
    def __init__(self, d, m, y):
        self.d, self.m, self.y = d, m, y

    def is_leap(self):
        return (self.y % 4 == 0 and self.y % 100 != 0) or (self.y % 400 == 0)

    def valid(self):
        if self.m < 1 or self.m > 12: return False
        ds = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        if self.is_leap(): ds[2] = 29
        return 1 <= self.d <= ds[self.m]

    def next(self):
        ds = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        if self.is_leap(): ds[2] = 29
        d, m, y = self.d + 1, self.m, self.y
        if d > ds[m]:
            d, m = 1, m + 1
            if m > 12: d, m, y = 1, 1, y + 1
        return f"{d:02d}/{m:02d}/{y:04d}"

try:
    s = input().split('/')
    d, m, y = map(int, s)
    dt = Date(d, m, y)
    if dt.valid(): print(dt.next())
    else: print("INVALID")
except:
    print("INVALID")
```

```python
#w8a3
class Point:
    def __init__(self, x, y):
        self.x, self.y = x, y

    def check(self):
        if self.x == 0 and self.y == 0: return "M la goc toa do"
        if self.x == 0: return "M nam tren truc tung"
        if self.y == 0: return "M nam tren truc hoanh"
        return ""

    def dist(self):
        return (self.x**2 + self.y**2)**0.5

x, y = map(float, input().split())
p = Point(x, y)
res = p.check()
if res: print(res)
print(f"{p.dist():.2f}")
```

```python
#w8a4
class Calc:
    def __init__(self, a, b):
        self.a, self.b = a, b
    def add(self): return self.a + self.b
    def sub(self): return self.a - self.b
    def mul(self): return self.a * self.b
    def div(self): return self.a / self.b if self.b != 0 else "Error"
    def pow(self): return self.a ** self.b
    def mod(self): return self.a % self.b

a, b = map(float, input().split())
c = Calc(a, b)
op = input()
if op == '+': print(c.add())
elif op == '-': print(c.sub())
elif op == '*': print(c.mul())
elif op == '/': print(c.div())
elif op == '^': print(c.pow())
elif op == '%': print(c.mod())
```

```python
#w8a5
class ShoppingCart:
    def __init__(self):
        self.items = []

    def addItem(self, n, p):
        self.items.append({"n": n, "p": p})

    def removeItem(self, n):
        self.items = [i for i in self.items if i["n"] != n]

    def isCartEmpty(self):
        return len(self.items) == 0

    def calculateTotal(self):
        return sum(i["p"] for i in self.items)

    def showCart(self):
        for i in self.items:
            print(f"{i['n']}: {i['p']}")

    def clearCart(self):
        self.items = []

# Ví dụ sử dụng:
cart = ShoppingCart()
cart.addItem("Sua", 20000)
cart.addItem("Banh", 15000)
cart.showCart()
print(cart.calculateTotal())
```

```python
#w8a6
class Rect:
    def __init__(self, w, h):
        self.w, self.h = w, h
    def area(self): return self.w * self.h
    def peri(self): return 2 * (self.w + self.h)
    def scale(self, k):
        self.w *= k
        self.h *= k

w, h, k = map(float, input().split())
r = Rect(w, h)
r.scale(k)
print(f"{r.area():.2f} {r.peri():.2f}")
```

```python
#w8a7
import math

class Fraction:
    def __init__(self, n, d):
        g = math.gcd(n, d)
        self.n = n // g
        self.d = d // g
        if self.d < 0:
            self.n, self.d = -self.n, -self.d

    def add(self, o):
        return Fraction(self.n * o.d + o.n * self.d, self.d * o.d)

    def sub(self, o):
        return Fraction(self.n * o.d - o.n * self.d, self.d * o.d)

    def mul(self, o):
        return Fraction(self.n * o.n, self.d * o.d)

    def div(self, o):
        return Fraction(self.n * o.d, self.d * o.n)

    def __str__(self):
        if self.d == 1: return str(self.n)
        return f"{self.n}/{self.d}"

# Cách sử dụng đơn giản:
line1 = list(map(int, input().split()))
line2 = list(map(int, input().split()))
f1 = Fraction(line1[0], line1[1])
f2 = Fraction(line2[0], line2[1])

print(f"Tong: {f1.add(f2)}")
print(f"Hieu: {f1.sub(f2)}")
print(f"Tich: {f1.mul(f2)}")
print(f"Thuong: {f1.div(f2)}")
#2 4
#1 4
```

```python
#w8a8
class Account:
    def __init__(self, name, bal):
        self.name, self.bal = name, float(bal)
    def deposit(self, v): self.bal += v
    def withdraw(self, v):
        if v <= self.bal: self.bal -= v
    def __str__(self): return f"{self.bal:.2f}"

s = input().split()
acc = Account(s[0], s[1])
n = int(input())
for _ in range(n):
    c, v = input().split()
    if c == "DEPOSIT": acc.deposit(float(v))
    else: acc.withdraw(float(v))
print(acc)
```

```python
#w8a9
class Student:
    def __init__(self, n):
        self.n, self.s = n, []
    def add(self, v): self.s.append(v)
    def avg(self): return sum(self.s)/len(self.s) if self.s else 0
    def rank(self):
        a = self.avg()
        if a >= 8: return "Excellent"
        if a >= 6.5: return "Good"
        if a >= 5: return "Average"
        return "Poor"

n = input()
st = Student(n)
m = int(input())
for _ in range(m):
    st.add(float(input().split()[1]))
print(f"{st.n} {st.avg():.2f} {st.rank()}")
```

```python
#w8a10
class Book:
    def __init__(self, t, a, y):
        self.t, self.a, self.y = t, a, y

class Library:
    def __init__(self): self.ls = []
    def add(self, b): self.ls.append(b)
    def count_a(self, a): return sum(1 for b in self.ls if b.a == a)
    def count_y(self, y): return sum(1 for b in self.ls if b.y == y)

lib = Library()
n = int(input())
for _ in range(n):
    cmd = input().split(' ', 1)
    if cmd[0] == "ADD":
        t, a, y = cmd[1].split(';')
        lib.add(Book(t, a, int(y)))
    elif cmd[0] == "COUNT": print(lib.count_a(cmd[1]))
    else: print(lib.count_y(int(cmd[1])))
```

```python
#w7a11
import math

class Point3D:
    def __init__(self, x, y, z):
        self.x, self.y, self.z = x, y, z

def get_normal(p1, p2, p3):
    # Vector v1 = P2 - P1, v2 = P3 - P1
    v1 = (p2.x - p1.x, p2.y - p1.y, p2.z - p1.z)
    v2 = (p3.x - p1.x, p3.y - p1.y, p3.z - p1.z)
    # Tich co huong [v1, v2] de tim vector phap tuyen
    nx = v1[1] * v2[2] - v1[2] * v2[1]
    ny = v1[2] * v2[0] - v1[0] * v2[2]
    nz = v1[0] * v2[1] - v1[1] * v2[0]
    return (nx, ny, nz)

def angle_between(n1, n2):
    # Tich vo huong
    dot = abs(n1[0]*n2[0] + n1[1]*n2[1] + n1[2]*n2[2])
    # Do dai vector
    mag1 = math.sqrt(n1[0]**2 + n1[1]**2 + n1[2]**2)
    mag2 = math.sqrt(n2[0]**2 + n2[1]**2 + n2[2]**2)
    
    cos_phi = dot / (mag1 * mag2)
    # Gioi han trong khoang [-1, 1] de tranh loi tinh toan so thuc
    cos_phi = max(-1, min(1, cos_phi))
    
    # Tra ve goc theo do
    return math.degrees(math.acos(cos_phi))

p = []
for _ in range(4):
    p.append(Point3D(*map(float, input().split())))

# Mat phang (ABC) tu p[0], p[1], p[2]
n1 = get_normal(p[0], p[1], p[2])
# Mat phang (BCD) tu p[1], p[2], p[3]
n2 = get_normal(p[1], p[2], p[3])

res = angle_between(n1, n2)
print(f"{res:.2f}")
```

```python
#w8a12
class Complex:
    def __init__(self, r, i):
        self.r, self.i = r, i
    def mul(self, o):
        return Complex(self.r*o.r - self.i*o.i, self.r*o.i + self.i*o.r)
    def __str__(self): return f"{self.r:.0f} {self.i:.0f}"

r1, i1 = map(float, input().split())
r2, i2 = map(float, input().split())
print(Complex(r1, i1).mul(Complex(r2, i2)))
```

```python
#w8a13
class Triangle:
    def __init__(self, a, b, c):
        self.a, self.b, self.c = a, b, c
    def area(self):
        a, b, c = self.a, self.b, self.c
        if a+b<=c or a+c<=b or b+c<=a: return "invalid"
        s = (a+b+c)/2
        return f"{(s*(s-a)*(s-b)*(s-c))**0.5:.2f}"

a, b, c = map(float, input().split())
print(Triangle(a, b, c).area())
```
