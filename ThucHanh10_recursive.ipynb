```python
#w9a8

vis = ['00']
moves = ""
nxt = [0, 0]
f = False
def Try(c, i, j):
    global f, vis
    if c == len(moves):
        return
    if moves[c] == 'R':
        nxt = [i, j+1]
    elif moves[c] == 'L':
        nxt = [i, j-1]
    elif moves[c] == 'D':
        nxt = [i+1, j]
    else:
        nxt = [i-1, j]
    
    tmp = ""
    tmp += str(nxt[0]) + str(nxt[1])
    if tmp in vis:
        f = True
    vis.append(tmp)
    Try(c+1, nxt[0], nxt[1])

def is_path_crossing(moves):
    global f
    Try(0, 0, 0)
    return f        
        
moves = input()
print(is_path_crossing(moves))
 
```

```python
#w9a9
def Median(nums):
    a, b, c, d, e = nums
    
    if a > b:
        a, b = b, a
    if c > d:
        c, d = d, c
    if a > c:
        a, b, c, d = c, d, a, b
    
    a = e
    
    if a > b:
        a, b = b, a
    if a > c:
        a, b, c, d = c, d, a, b
    if b < c:
        return b
    return c
```

```python
#w9a10
import bisect

def findRadius(houses, heaters):
    heaters.sort()
    ans = 0
    for h in houses:
        idx = bisect.bisect_left(heaters, h)
        
        dist_right = heaters[idx] - h if idx < len(heaters) else float('inf')
        dist_left = h - heaters[idx - 1] if idx > 0 else float('inf')
        
        ans = max(ans, min(dist_left, dist_right))
    return ans

a = list(map(int, input().split()))
b = list(map(int, input().split()))
print(findRadius(a, b))
```
