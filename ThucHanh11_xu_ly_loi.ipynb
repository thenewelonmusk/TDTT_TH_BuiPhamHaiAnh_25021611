```python
#w10a3

def is_happy(n):
    seen = set()
    while n != 1:
        cur = 0
        for i in str(n):
            cur += int(i)**2
        n = cur
        
        if n in seen:
            return 'NO'
        seen.add(n)
    return 'YES'

try:
    a = int(input())
    
    print(is_happy(a))
    if a < 0:
        print('vui long nhap so nguyen duong')
except ValueError:
    print('vui long nhap dung kieu du lieu')
```

```python
#w10a10

def valid(a):
    seen = set()
    cur = 0
    for i in a:
        if i in seen:
            return 'Mang khong hop le'
        seen.add(i)
        cur += i
    return cur

a = list(map(int, input().split()))
print(valid(a))
```

```python
#w10a11
import re

def check_regex():
    try:
        t_str = input().strip()
        if not t_str:
            return
        t = int(t_str)
        
        for _ in range(t):
            s = input().strip()
            try:
                re.compile(s)
                print("True")
            except re.error:
                print("False")
                
    except ValueError:
        print("Vui lòng nhập một số nguyên cho T.")

check_regex()
```
