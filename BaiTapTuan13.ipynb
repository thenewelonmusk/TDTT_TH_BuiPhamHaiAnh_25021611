```python
#w11a1
def reverse(path: str) -> int:
    with open(path, 'r') as f:
        s = f.read().strip()
    ans = int(s[::-1])
    return ans

s = input()
print(reverse(s))
```

```python
#w11a2
def can_win(path: str) -> bool:
    with open(path, 'r') as f:
        n = int(f.read().strip())
    ans = n % 4 != 0
    return ans

s = input()
print(can_win(s))
```

```python
#w11a4
def check(p1: str, p2: str) -> bool:
    def get_c(p):
        with open(p, 'r') as f:
            s = f.read().lower()
        return sorted([c for c in s if c.isalnum()])
    
    ans = get_c(p1) == get_c(p2)
    return ans

f1 = input()
f2 = input()
print(check(f1, f2))
```

```python
#w11a10
def getMoney(p: str) -> list[int]:
    with open(p, 'r') as f:
        a = [int(x) for x in f.read().split()]
    return a

def rob(a: list[int]) -> int:
    p, c = 0, 0
    for x in a:
        p, c = c, max(p + x, c)
    ans = c
    return ans

p = input()
a = getMoney(p)
print(rob(a))
```

```python
#w11a12
def maximumProduct(p: str) -> int:
    with open(p, 'r') as f:
        f.readline()
        a = sorted([int(x) for x in f.read().split()])
    
    ans = max(a[-1] * a[-2] * a[-3], a[0] * a[1] * a[-1])
    return ans

p = input()
print(maximumProduct(p))
```

```python
#w11a13
def moveZeroes(fileName: str) -> list:
    with open(fileName, 'r') as f:
        f.readline()
        a = [int(x) for x in f.read().split()]
    
    ans = [x for x in a if x != 0]
    ans += [0] * (len(a) - len(ans))
    return ans

fn = input()
print(moveZeroes(fn))
```

```python
#w11a14
def averageTime(p: str):
    with open(p, 'r') as f:
        a = [float(x.strip().split(',')[1]) for x in f.readlines()[1:]]
    ans = round(sum(a) / len(a), 2)
    return ans

p = input()
print(averageTime(p))
```

```python
#w11a15
def findMovies(p: str):
    with open(p, 'r') as f:
        a = f.readlines()[1:]
    
    ans = []
    for x in a:
        b = [i.strip() for i in x.split(',')]
        if float(b[3]) > 8.0 and b[2].lower() != "boring":
            ans.append(b[1])
            
    return ans if ans else "All are boring!"

p = input()
print(findMovies(p))
```

```python
#w11a16
import math

p = input()
with open(p, 'r') as f:
    L = [line.strip() for line in f if line.strip()]

for i in range(0, len(L), 2):
    t = L[i]
    v = L[i+1]
    if "SQUARE" in t:
        s = float(v.split(":")[1])
        ans = 4 * s
    elif "RECTANGLE" in t:
        b = v.split()
        w = float(b[1])
        h = float(b[3])
        ans = 2 * (w + h)
    elif "CIRCLE" in t:
        r = float(v.split(":")[1])
        ans = 2 * math.pi * r
    print(f"{ans:.2f}")
```

```python
#w11a17
def productExceptSelf(p: str) -> list[int]:
    with open(p, 'r') as f:
        a = [int(x) for x in f.read().split()]
    
    n = len(a)
    ans = [1] * n
    
    l = 1
    for i in range(n):
        ans[i] = l
        l *= a[i]
        
    r = 1
    for i in range(n - 1, -1, -1):
        ans[i] *= r
        r *= a[i]
        
    print(ans)
    return ans

p = input()
productExceptSelf(p)
```

```python
#w11a18
import math

def estimatedTime(p: str):
    with open(p, 'r') as f:
        a = [line.split() for line in f]
    
    x1, y1 = float(a[0][2]), float(a[0][4])
    x2, y2 = float(a[1][2]), float(a[1][4])
    v = float(a[2][1])
    
    d = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    ans = round(d / v, 2)
    return ans

p = input()
print(estimatedTime(p))
```

```python
#w11a20
def imageSmoother(path: str):
    with open(path, 'r') as f:
        m = int(f.readline().strip())
        a = [[int(x) for x in f.readline().split()] for _ in range(m)]

    ans = [[0] * m for _ in range(m)]

    for r in range(m):
        for c in range(m):
            s, count = 0, 0
            for dr in [-1, 0, 1]:
                for dc in [-1, 0, 1]:
                    nr, nc = r + dr, c + dc
                    if 0 <= nr < m and 0 <= nc < m:
                        s += a[nr][nc]
                        count += 1
            ans[r][c] = s // count

    for row in ans:
        print(*(row))

p = input()
imageSmoother(p)
```
